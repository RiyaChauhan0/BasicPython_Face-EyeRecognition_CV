```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('demo.jpg')
model_face = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
model_eye = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
demo_face = model_face.detectMultiScale(img,1.1,10)
demo_eye = model_eye.detectMultiScale(img,1.05,20)
count = 1
for (x,y,w,h) in demo_face:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
  cv2.putText(img,f"Person{count}",(x,y),cv2.FONT_HERSHEY_PLAIN,2,(0,0,0),2)
  roi = img[y:y+h,x:x+w] # Extracting required faces
  cv2_imshow(roi)
  cv2.imwrite(f'Face {count}.jpg',roi)
  count+=1
for (x,y,w,h) in demo_eye:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),5)
cv2_imshow(img)
```
